## 小作业1: 哈夫曼树可视化

根据26个字母及其相应的权重值，设计算法来对其构建一个哈夫曼树，并计算其对应的编码长度。要求实现以下功能：

1. 构建哈夫曼树：根据给定的字符及其权重值，构建出对应的哈夫曼树，将生成的哈夫曼树进行可视化展示，使得树的层次结构和每个字符的编码直观可见。
2. 计算编码长度：对于构建出的哈夫曼树，计算每个字符的编码长度，即根节点到该字符叶子节点的路径长度之和。路径长度定义为从根节点到某个节点的边的数量。（完成）
3. 输出编码表：根据哈夫曼树，生成字符到对应编码的映射表。将每个字符与其对应的哈夫曼编码打印输出。（完成）
4. 设计界面输入任意字符串，输出对应的哈夫曼代码。(完成)

```python
import heapq as hq #python经典的可以实现优先队列的东西
import tkinter as tk
from tkinter import messagebox
from graphviz import Digraph
from IPython.display import Image
```

```python
# treenode类
class treenode:
    def __init__(self,letter = None,code = None,weight = None,codelength = None):
        self.letter = letter
        self.code = code
        self.codelength = codelength
        self.weight = weight
        self.Left = None
        self.Right = None
    def __lt__(self,other):
        return self.weight < other.weight #lt是一个Python的特性。可以告诉Python怎么“比较”两个对象


    
    
        
```

```python
class Interface:
    def __init__(self, letter2code):
        self.letter2code = letter2code
        
        # 主窗口
        self.window = tk.Tk()#这是一个窗口对象
        self.window.title("字符串编码器")#标题
        self.window.geometry("500x200")#大小
        
        # 创建和布局这个可视化界面
        tk.Label(self.window, text="字符串编码转换器").pack(pady=10)#pady可以控制上下间距10
        tk.Label(self.window, text="输入字符串:").pack()#写了pack才能渲染！不写不行
        self.entry_input = tk.Entry(self.window, width=40)
        self.entry_input.pack(pady=5)
        tk.Button(self.window, text="编码", command=self._encode).pack(pady=5) #command=self._encode 唤起下面的encode模块
        
        self.label_output = tk.Label(self.window, text="编码结果:", font=("Arial", 12))
        self.label_output.pack(pady=10)

    def _encode(self):
        input_str = self.entry_input.get()#获取输入
        encoded = encode_string(input_str,letter2code)#这个函数下面有写
        
        self.label_output.config(text="编码结果: " + ''.join(encoded))
        self.entry_input.delete(0, tk.END)

    def run(self):
        """启动界面主循环"""
        self.window.mainloop()
    
```

```python
letter_frequency = [
    ['A', 0.0819], ['B', 0.0147], ['C', 0.0383], ['D', 0.0391], ['E', 0.1225],
    ['F', 0.0226], ['G', 0.0171], ['H', 0.0457], ['I', 0.0710], ['J', 0.0041],
    ['K', 0.0014], ['L', 0.0377], ['M', 0.0334], ['N', 0.0706], ['O', 0.0726],
    ['P', 0.0289], ['Q', 0.0009], ['R', 0.0685], ['S', 0.0636], ['T', 0.0941],
    ['U', 0.0258], ['V', 0.0109], ['W', 0.0159], ['X', 0.0021], ['Y', 0.0158],
    ['Z', 0.0008]
] # 这部分的确用了ai，因为手打字打那个表格很费力，我让豆包帮我把表格抄下来了
```

```python
nodes = [] #创建一个nodes表
for letter,freq in letter_frequency:
    node = treenode(letter,None,freq)
    nodes.append(node)
```

### 功能1的实现：
1. 构建树
2. 可视化展示

```python
#完成建构哈夫曼树的任务
def build_tree(nodes):
    hq.heapify(nodes)#变成优先队列
    while len(nodes)>1: #在只剩下一个节点的时候才停止
        left = hq.heappop(nodes)#更小的在左侧
        right = hq.heappop(nodes)
        new = treenode(weight = (left.weight + right.weight))#新节点
        new.Left = left
        new.Right = right #给new分配孩子们

        hq.heappush(nodes,new) #推入，之后优先队列会自己排列好
        
    return nodes[0]#根节点
    
```

```python
# 给字母分配编码的功能
def assign_codes(node,current_code = "",letter2code = None):  #刚开始的时候l2c肯定是None
    if letter2code is None:
        letter2code = {} #同时记录一下
        
    if node is not None:#不能是空节点
        if node.letter is not None: #只给叶子节点，也就是有实际字母的节点赋值！
            node.code = current_code
            node.codelength = len(current_code)
            letter2code[node.letter] = current_code
        assign_codes(node.Left,current_code + "0",letter2code)#递归，往左走加0，往右走加1
        assign_codes(node.Right,current_code + "1",letter2code)#可变对象通过引用传递，因此递归调用的每一层操作的是同一个字典对象。
    return letter2code
```

```python
#输出编码表和长度表
def output_code(letter2code):
    for letter,code in letter2code.items():#可以遍历的对象
        print(f"{letter} 的编码是{code},编码长度为{len(code)}")
```

```python
#编码功能
def encode_string(input_str,letter2code):
    encoded = "" #已被输出的串
    input_str = input_str.upper() #大写
    for char in input_str:
        if char in letter2code:
            encoded += letter2code[char]
        else:
            encoded += "?" #无法编码就暂用？代替
    return encoded
```

### 功能2,3的实现

同时打印编码表和长度

```python
tree = build_tree(nodes)
letter2code = assign_codes(tree)
output_code(letter2code)
```

```python
def visualize_huffman(root, filename="huffman_tree"):
    #可视化使用了一个叫graphviz的库
    dot = Digraph(format='png')#有向图
    
    def add_nodes(node):
        if node is None:
            return
        
        # 如果是实际的字母就把字母打印出来否则就是空
        if node.letter is not None:
            label = f"{node.letter}\nweight={node.weight}"
        else:
            label = f"{''}\nweight={node.weight}"
        if node.code is not None:
            label += f"\ncode={node.code}"
        
        dot.node(str(id(node)), label) #前面先获取node的内存地址作为id，防止节点重复等,然后label是标签，这一步把这个节点实际地添加进来
        
        # 递归
        if node.Left is not None:
            add_nodes(node.Left)
            dot.edge(str(id(node)), str(id(node.Left)), label="0") # 添加边
        if node.Right is not None:
            add_nodes(node.Right)
            dot.edge(str(id(node)), str(id(node.Right)), label="1")
    
    add_nodes(root)
    return Image(dot.render(filename, format='png', cleanup=True))#渲染，把图片渲染上去
```

```python
visualize_huffman(tree)
```
